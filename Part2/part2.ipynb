```python
#Part 2
#1. Load cleaned_data.csv from Part 1. 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge   
from sklearn.metrics import (
    mean_squared_error,
    r2_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    roc_curve, 
    confusion_matrix,
    classification_report,
)

df = pd.read_csv("BHP_cleaned.csv")
df.head()




```

```python
print(df.columns)
targetReg = df["price"]
classificationTarget = (targetReg > targetReg.median()).astype(int)
Matrix = df.drop("price", axis=1)

print("Feature Matrix:", Matrix.shape)
print("Regression Target:", targetReg.shape)
print("Classification Target:", classificationTarget.value_counts())
```

```python
#2. Encode categorical columns:
print(Matrix.dtypes)
Matrix = pd.get_dummies(Matrix, drop_first=True)
print(Matrix.shape)
```

```python
#3. Leak-free train-test split and scaling:
Matrix_train_reg, Matrix_test_reg, targetReg_train, targetReg_test = train_test_split(
    Matrix, targetReg, test_size=0.2, random_state=42
)
Matrix_train_clf, Matrix_test_clf, targetClf_train, targetClf_test = train_test_split(
    Matrix, classificationTarget, test_size=0.2, random_state=42
)

scaler_reg = StandardScaler()
Matrix_train_reg_scaled = scaler_reg.fit_transform(Matrix_train_reg)
Matrix_test_reg_scaled = scaler_reg.transform(Matrix_test_reg)      

scalerClassification = StandardScaler()
Matrix_train_clf_scaled = scalerClassification.fit_transform(Matrix_train_clf)
Matrix_test_clf_scaled = scalerClassification.transform(Matrix_test_clf)
print("Scaling complete.")
```

```python
#4.Regression model — Linear Regression:
linearRegModel = LinearRegression()
linearRegModel.fit(Matrix_train_reg_scaled, targetReg_train)    
predictions_reg = linearRegModel.predict(Matrix_test_reg_scaled)

mse = mean_squared_error(targetReg_test, predictions_reg)
r2 = r2_score(targetReg_test, predictions_reg)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
coefficients = pd.DataFrame({
    "Feature": Matrix.columns,
    "Coefficient": linearRegModel.coef_
})

coefficients["Absoulte"] = coefficients["Coefficient"].abs()
coefficients_sorted = coefficients.sort_values(by="Absoulte", ascending=False)
print(coefficients.head(10))
print("\nTop 3 Important features with highest absolute coefficients:")
print(coefficients_sorted.head(3))

```
